```python
#르벤슈타인 라이브러리 설치
!pip install python-Levenshtein

#코랩으로 이용하기 때문에 파일을 직접 업로드하기
from google.colab import files
uploaded = files.upload()  # 업로드 창이 뜸

#업로드한 파일 자동으로 가져오기
file_name = list(uploaded.keys())[0]  # 예: 'ChatbotData.csv'

#CSV파일 읽기 라이브러리 불러오기
import pandas as pd
#르벤슈타인 거리 계산 함수 라이브러리 불러오기
import Levenshtein

#클래스 정의하기
class SimpleChatBot:
    def __init__(self, filepath):
        self.questions, self.answers = self.load_data(filepath) #파일 경로받고 질문/답변 데이터 불러오기

    def load_data(self, filepath): #질문, 답변 리스트 분리해 반환하기
        data = pd.read_csv(filepath)
        questions = data['Q'].tolist()
        answers = data['A'].tolist()
        return questions, answers

    def find_best_answer(self, input_sentence): #내가 입력한 문장과 비슷한 질문 찾고, 그에 해당하는 답변 반환하기
        shortest_distance = float('inf')  #가장 짧은 거리 초기값
        best_match_index = -1  #가장 유사한 질문의 인덱스 저장 변수

        for i, question in enumerate(self.questions): #모든 질문을 돌며 내가 입력한 르벤슈타인 거리 비교
            distance = Levenshtein.distance(input_sentence, question)  #르벤슈타인거리 함수 사용 Levenshtein.distance()
            if distance < shortest_distance:
                shortest_distance = distance
                best_match_index = i

        return self.answers[best_match_index]  #가장 가까운 질문 인덱스를 이용해 답변 반환

#챗봇 실행하기
chatbot = SimpleChatBot(file_name)

print("✅ 챗봇을 시작합니다. '종료'를 입력하면 챗봇이 끝납니다♡.")

while True:
    user_input = input("You: ")
    if user_input.strip().lower() == "종료":
        print("Chatbot: 안녕히 가세요!")
        break
    response = chatbot.find_best_answer(user_input)
    print("Chatbot:", response)
```
